### Making function that will check the grammar and spelling

```python
def check(my_text):
    # making object 
    tool = language_tool_python.LanguageToolPublicAPI('en-US')
        
    # performing spell check
    text = tool.check(my_text)
    
    # making some list to store information about our text
    start_position = []         # start position of the error
    end_position = []           # end position of the error
    error = []                  # which word has error
    correction = []             # correction of error
    
    for i in text:
        start_position.append(i.offset)
        end_position.append(i.errorLength + i.offset)
        error.append(my_text[i.offset : i.errorLength + i.offset])
        try:
            correction.append(i.replacements[0])
        except:
            correction.append(error[-1])
        
    # creating copy of our text in a list form
    my_text_new = list(my_text)
    
    for i in range(len(start_position)):
        x=0
        for n in range(len(my_text_new)):
            if (x==0):
                my_text_new[start_position[i]] = correction[i]
                x=1
            if (n > start_position[i] and n < end_position[i] ):
                my_text_new[n] = ''
                
    my_text_new = "".join(my_text_new)
    # calculating the accuracy
    number_of_words = len(my_text.split())
    number_of_error_words = len(error)
    
    accuracy = (number_of_error_words/number_of_words) *100
    accuracy = 100 - accuracy
    return (my_text_new , accuracy)
```

```python
(new_text ) = df['text'].apply(check)    
```
